```python
#Calculate Avg Temperature from temp column to draw a line graph
daily_avg_temp= df.groupby('dteday')['temp'].mean()

# Plot the line graph
plt.figure(figsize=(12,6))
plt.plot(daily_avg_temp.index, daily_avg_temp.values, marker='o', linestyle='-', color='b')

for x,y in zip(daily_avg_temp.index, daily_avg_temp.values):
    plt.text(x,y, f'{y:.2f}', fontsize=9, ha='right', va='bottom', color='black')
    
plt.xlabel('Date')
plt.ylabel('Average Temperature')
plt.title('Daily Average Temperature in January')
plt.xticks(rotation=45)
plt.grid()
```

```python
# Distribution of temperature throuhout the month
plt.figure(figsize=(8,5))
sns.histplot(df['temp'], bins=20, kde=True)
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Distribution of temperature')
plt.show()
```

# Histogram- It is used to show the distribution of the data in time intervals.

```python
# Bar Graph for Average temperature by weekday
weekday_mapping= {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
avg_temp_by_weekday= df.groupby('weekday', as_index= False)['temp'].mean()
avg_temp_by_weekday['weekday']= avg_temp_by_weekday['weekday'].map(weekday_mapping)
avg_temp_by_weekday=avg_temp_by_weekday.drop_duplicates(subset=['weekday'])
weekday_order= ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
avg_temp_by_weekday['weekday']=pd.Categorical(avg_temp_by_weekday['weekday'],categories=weekday_order, ordered=True)
avg_temp_weekday=avg_temp_by_weekday.sort_values('weekday')

plt.figure(figsize=(10,5))
ax= sns.barplot(x='weekday',y='temp', data= avg_temp_by_weekday, palette='coolwarm')
for index, value in enumerate(avg_temp_by_weekday['temp']):
    ax.text(index, value, f"{value:.2f}", ha='center', va= 'bottom',fontsize=10)
plt.xlabel('Weekday')
plt.ylabel('Average Temperature')
plt.title('Average Temperature by Weekday')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
numerical_cols = merged_123.select_dtypes(include=[np.number])
Q1 = numerical_cols.quantile(0.25)
Q3 = numerical_cols.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = (numerical_cols < lower_bound) | (numerical_cols > upper_bound)

outlier_values = numerical_cols[outliers]
print(outlier_values.stack())
```

# Outliers- Outliers are those values which are greater than the upper bound value and lesser than lower bound value.
# Outliers are calculated with IQR method.
# Outlier values with respected columns are printed, total 252 outliers in all columns combined.

```python
outlier_counts = outliers.sum()
print(outlier_counts)
```

```python
transformed_casual= merged_123['casual_log']=np.log1p(merged_123['casual'])
transformed_registered= merged_123['registered_log']=np.log1p(merged_123['registered'])  
transformed_cnt= merged_123['cnt_log']=np.log1p(merged_123['cnt'])

print("Original Skewness Casual: ", merged_123['casual'].skew())
print("\nTransformed Skewness Casual: ", transformed_casual.skew())
print("\nOriginal Skewness Registered: ", merged_123['registered'].skew())
print("\nTransformed Skewness Registered: ", transformed_registered.skew())
print("\nOriginal Skewness Cnt: ", merged_123['cnt'].skew())
print("\nTransformed Skewness Cnt: ", transformed_cnt.skew())
```

# Columns casual, registered and cnt are transformed through log transformation to adjust skewness.
# The skewness of all the three columns is reduced by the transformation.

```python
fig, axes= plt.subplots(3,2, figsize=(12,12))
columns=['casual', 'registered', 'cnt']
transformed_data= [transformed_casual, transformed_registered, transformed_cnt]

for i, col in enumerate(columns):
    sns.histplot(merged_123[col], bins=30, kde=True, ax=axes[i,0], color='blue')
    axes[i,0].set_title(f'Original {col} Distribution')

    sns.histplot(transformed_data[i], bins=30, kde=True, ax=axes[i,1], color='green')
    axes[i,1].set_title(f'Transformed {col} (log) Distribution')

plt.tight_layout()
plt.show()
```
